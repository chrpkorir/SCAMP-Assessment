# Visualize data


1. Best selling books of the decade

```python
df[['Name','Price']].nlargest(10,'Price').plot.bar(x='Name',y='Price',color='green')
plt.title('Top 10 books with highest price 2009-2019',fontsize=16)
plt.xlabel('Name of the Books',color='blue',fontsize=15)
plt.ylabel('Price',color='blue',fontsize=15)
plt.show()  
```

2. Top 10 authors according to the reviews

```python
df[['Author','Reviews']].nlargest(10,'Reviews').plot.bar(x='Author',y='Reviews',color='coral')
plt.title('Top 10 Authors according to reviews 2009-2019',size=15)
plt.xlabel('Authors',size=15,color='blue')
plt.ylabel('Number of reviews',size=15, color='blue')
plt.show()
```

3. The percentage of finction to non-finction


```python
df['Genre'].value_counts()
```

```python

slices = [310, 240]
activities = ['Non Fiction', 'Fiction']
colors = ['r', 'g']
plt.pie(slices, labels=activities, colors=colors, startangle=90, autopct='%.1f%%')
plt.title('Percentage of each genre')
plt.show()
```

4. Trends over the years

```python
df[['Author','Price']].nlargest(10,'Price').plot.bar(x='Author',y='Price',color='yellow')
plt.title('Top 10 Authors according to the price of the books 2009-2019',size=15)
plt.xlabel('Authors',size=15,color='blue')
plt.ylabel('Number of reviews',size=15, color='blue')
plt.show()
```
